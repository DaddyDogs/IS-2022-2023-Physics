```python
import math
import numpy as np
import matplotlib.pyplot as plt
exp = 2.71828
g = 9.81
length = 0.613125
m = 12
k = 0.3 #коэффициент упругости стали
friction_coeff = 0.8 #коэффициент сопротивления
damping_factor = friction_coeff / (2 * m)
natural_frequency = math.sqrt(k / m)
speed = 20
amplitude_0 = math.sqrt(speed ** 2 / natural_frequency ** 2)
t = np.arange(0, 160, 10 ** -3)
y = [0] * t.size
```

```python
sqrtt = math.sqrt(4 * natural_frequency ** 2 - damping_factor ** 2)
for i in range(t.size):
    y[i] = 1 / 2 * (exp ** (-t[i] * damping_factor / 2))  * amplitude_0 * (np.cos(sqrtt * t[i]) + np.sin(sqrtt * t[i]))
```

```python
fig, ax = plt.subplots()
ax.plot(t, y, color='cyan')
x = [0] * t.size
ax.plot(t, x, color='gray', linestyle='dashed')
ax.set_xlabel('time')
plt.show()
```

```python
sqrtt = math.sqrt(4 * natural_frequency ** 2 - damping_factor ** 2)
force_frequency = 0.25
particular_solution1 = amplitude_0 * (natural_frequency ** 2 - force_frequency ** 2) / (damping_factor ** 2 * force_frequency ** 2 + (natural_frequency ** 2 - force_frequency ** 2) ** 2)
particular_solution2 = amplitude_0 * damping_factor * force_frequency / (damping_factor ** 2 * force_frequency ** 2 + (natural_frequency ** 2 - force_frequency ** 2) ** 2)
for i in range(t.size):
    y[i] = 1 / 2 * (exp ** (-t[i] * damping_factor / 2))  * amplitude_0 * (np.cos(sqrtt * t[i]) + np.sin(sqrtt * t[i])) + particular_solution1 * np.cos(force_frequency * t[i]) + particular_solution2 * np.sin(force_frequency * t[i])
```

```python
fig, ax = plt.subplots()
ax.plot(t, y, color='cyan')
x = [0] * t.size
ax.plot(t, x, color='gray', linestyle='dashed')
ax.set_xlabel('time')
plt.show()
```
